```python
# Pro hlavní stránku 
def distance_finder(data,step):
    '''
    Tato fuknce má za cíl nalézt najeté kilometry automobilu
    '''
    for i in range(0,len(re.findall(',',data[step].text))+1,1):
        if re.split(',',data[step].text)[i].replace("\xa0","")[-2:] in ['km','KM','Kilometrů','Km']:
            KM = re.split(',',data[step].text)[i].replace("\xa0","").replace('km','')
            break
        else:
            KM = np.nan
    return float(KM)


typy_paliva = ['Nafta','nafta','benzín','Benzín','CNG+benzín','CNG + benzín','CNG+nafta','CNG + nafta',"Hybridní","LPG+benzín",'LPG + benzín']
def fuel_finder(data,step):
    '''
    Tato fuknce má za cíl nalézt druh paliva
    '''
    for i in range(0,len(re.findall(',',data[step].text))+1,1):
        if re.split(',',data[step].text)[i].replace(" ","") in typy_paliva:
            fuel = re.split(',',data[step].text)[i].replace(" ","")
            break
        else:
            fuel = 'NA'
    return str(fuel)
        

def prevodovka(data,step):
    '''
    Tato fuknce má za cíl nalézt typ převodovky
    '''
    for i in range(0,len(re.findall(',',data[step].text))+1,1):
        if re.split(',',data[step].text)[i].replace(" ","") in ['Automatická','automatická','manuální','Manuální']:
            prevodovka = re.split(',',data[step].text)[i].replace(" ","")
            break
        else:
            prevodovka = 'NA'
    return str(prevodovka)

def get_year(data,step):
    '''
    Tato fuknce má za cíl nalézt informaci od kdy je auto v provozu
    '''
    year = data[step].text[0:4]
    return float(year)
    
    
def get_url(data):
    '''
    Tato fuknce má za cíl nalézt URL daného automobilu pro získání dalších informací 
    '''
    url_local = re.search('c-item__link" href="(.+?)" rel="noreferrer',str(data)).group(1)
    return url_local

def get_price(data,step):
    '''
    Tato fuknce má za cíl nalézt cenu daného automobilu 
    '''    
    price_cleaned = data[step].text.replace("\xa0","")[:-2]
    return price_cleaned

def get_currency(data,step):
    '''
    Tato fuknce má za cíl nalézt v jaké měně je cena
    '''    
    currency = data[step].text.replace("\xa0","")[-2:]
    return currency


def EURO(data):
    '''
    Tato fuknce má za cíl nalézt EURO kritérium daného automobilu 
    '''    
    for i in range(0,len(data)+1,1):
        try:
            if data[i][0:4] == 'EURO':
                EURO = data[i]
                break
            else:
                EURO="NA"
        except:
            EURO = "NA"
    return EURO


def KW(data):
    '''
    Tato fuknce má za cíl nalézt KW automobilu 
    '''  
    for i in range(0,len(data)+1,1):
        try:
            if data[i][4:6] in ['kw','KW','kW','Kw']:
                kw_finder = data[i][0:3]
                break
            if data[i][3:5] in ['kw','KW','kW','Kw']:
                kw_finder = data[i][0:3]
                break
            else:
                kw_finder = '<NA>'
        except:
            kw_finder = '<NA>'
    return kw_finder


def ccm(data):
    '''
    Tato fuknce má za cíl nalézt KW automobilu 
    '''  
    for i in range(0,len(data)+6,1):
        try:
            if data[i][-3:] in ['ccm','cCm','CCM','cCm','Ccm']:
                ccm_finder = data[i].replace("\xa0","").replace('ccm','')
                break
            else:
                ccm_finder = '<NA>'
        except:
            ccm_finder = '<NA>'
    
    return ccm_finder

            
```

```python
def price_qrow(row):
    if row['Usage_time'] <= 1:
        return 0.03
    elif (row['Usage_time'] > 1 and row['Usage_time'] <= 5):
        return 0.07
    elif row['Usage_time'] > 5:
        return 0.10
```

```python
class Car:
    def __init__(self, Mileage, KW,cmm,Usage_time,Gearbox,Euro_emissions,Fuel_type):
        self.Mileage = Mileage
        self.KW = KW
        self.cmm = cmm
        self.Usage_time = Usage_time
        self.Gearbox = Gearbox
        self.Euro_emissions = Euro_emissions 
        self.Fuel_type = Fuel_type
        
    def exogen_val(self):
        constant = 1
        if self.Gearbox == 'Manuální':
            gearbox = 1
        elif self.Gearbox != 'Manuální':
            gearbox = 0
        if self.Euro_emissions =='EURO 2':
            EURO_2  = 1
            EURO_3  = 0 
            EURO_4  = 0 
            EURO_5  = 0 
            EURO_6  = 0 
        elif self.Euro_emissions =='EURO 3':
            EURO_2  = 0
            EURO_3  = 1 
            EURO_4  = 0 
            EURO_5  = 0 
            EURO_6  = 0 
        elif self.Euro_emissions =='EURO 4':
            EURO_2  = 0
            EURO_3  = 0 
            EURO_4  = 1 
            EURO_5  = 0 
            EURO_6  = 0 
        elif self.Euro_emissions =='EURO 5':
            EURO_2  = 0
            EURO_3  = 0 
            EURO_4  = 0 
            EURO_5  = 1 
            EURO_6  = 0 
        elif self.Euro_emissions =='EURO 6':
            EURO_2  = 0
            EURO_3  = 0 
            EURO_4  = 0 
            EURO_5  = 0 
            EURO_6  = 1
        elif self.Euro_emissions not in ['EURO 1','EURO 2','EURO 3','EURO 4','EURO 5','EURO 6']:
            EURO_2  = 0
            EURO_3  = 0 
            EURO_4  = 0 
            EURO_5  = 0 
            EURO_6  = 0
#         return sum((constant,self.Mileage**(1/3),self.KW,self.cmm,self.Usage_time**(1/3),gearbox,EURO_2,EURO_3,EURO_4,EURO_5,EURO_6)*res_naf.params)
        return constant,self.Mileage**(1/3),self.KW,self.cmm,self.Usage_time**(1/3),gearbox,EURO_2,EURO_3,EURO_4,EURO_5,EURO_6

    def pred(self):
        if self.Fuel_type == 'Nafta':
            p = sum(res_naf.params*(Car1.exogen_val()))
        elif self.Fuel_type == 'Benzín':
            p = sum(res_ben.params*(Car1.exogen_val()))
        return p
    
```
